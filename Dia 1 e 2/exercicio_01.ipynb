### Carregar uma imagem em preto e braco:
    - Img = data.page()

```python
from skimage import data
from skimage import io 
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

img = data.page()
# img = data.astronaut()
# img = io.imread("Brasil x Japão.webp")

print("Dimensôes: ", img.shape)
print("Tipo: ", img.dtype)

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title("Imagem da biblioteca (camera)")
plt.axis("off")
plt.show()
```

### Inverter as cores

```python
imagem_invertida = 255-img

print("Dimensôes: ", imagem_invertida.shape)
print("Tipo: ", imagem_invertida.dtype)

plt.imshow(imagem_invertida, cmap='gray', vmin=0, vmax=255)
plt.title("Imagem Invertida")
plt.axis("off")
plt.show()
```

### Rotacionar 90°

```python
img_rot = np.rot90(img)

plt.imshow(img_rot, cmap='gray', vmin=0, vmax=255)
plt.title("Imagem Rotacionada")
plt.axis("off")
plt.show()
```

### Versão Espelhada

```python
img_espelhada = np.fliplr(img)

plt.imshow(img_espelhada, cmap='gray', vmin=0, vmax=255)
plt.title("Imagem Espelhada")
plt.axis("off")
plt.show()
```

### Aumentar o Brilho

```python
img_clara = np.clip(img + 0, 0, 100)
plt.imshow(img_clara, cmap='gray')
plt.title("Brilho Aumentado")
plt.axis("off")
plt.show()
```
